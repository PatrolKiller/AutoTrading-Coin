```python
from trading import trade

# 시작 날짜 입력
startyear, startmonth, startday = 2021, 1, 1
# 종료 날짜 입력
endyear, endmonth, endday = 2021, 3, 1
# 코인이름 입력 ex)'KRW-ETH'
ticker = 'KRW-ETH'
# 코인차트 기준봉
interval = 'minute15'
# 타임델타 기준시간
minutes = 15

day_total, assets_total = trade(startyear, startmonth, startday, endyear, endmonth, endday, interval, minutes, ticker)
```

```python
import matplotlib.pyplot as plt

ticker = 'KRW-ETH'
plt.style.use('fivethirtyeight')
plt.figure(figsize=(20, 11))
plt.plot(day_total, assets_total, lw = 1.2)
plt.ylabel('Valuation asset')
plt.title('ticker: {}   interval: {}'.format(ticker, interval))
```
